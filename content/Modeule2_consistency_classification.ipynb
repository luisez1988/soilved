Based on Brinas, Merediths, and Ashly's notes

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import Soilved as soil
```

```python
#Define GSD data
soil.GSD_data = pd.read_csv('./data/mod2_Brinas_example.csv')
soil.GSD_data.head()
```

```python
#Plot Grain size distribution
plt.figure(figsize=(8,4))


#Add notable sieve sizes in vertical lines of light red color
for sieve_size in [75, 50, 25, 25.4/2, 4.75, 2, 0.425, 0.15, 0.075, 0.005]:
    plt.axvline(x=sieve_size, color='lightcoral', linestyle='--', linewidth=0.8)
#add annotations for notable sieve sizes in in and sieve number at the top margin of the plot
sieve_annotations = {
    75: '3 in',
    50: '2 in',
    25: '1 in',
    10: '1/2 in',
    4.75: 'No. 4',
    2: 'No. 10',
    0.425: 'No. 40',
    0.15: 'No. 100',
    0.075: 'No. 200',
}
for sieve_size, label in sieve_annotations.items():
    plt.text(sieve_size, 105, label, rotation=90, verticalalignment='bottom', horizontalalignment='center', color='darkred', fontsize=8)

plt.plot(soil.GSD_data['d'], soil.GSD_data['Passing'], marker='o')
plt.xscale('log')
#make plot span from 0.001 mm to 100 mm
plt.xlim(100, 0.001)
#plt.gca().invert_xaxis()
#Make format of x axis be numbers such 0.1, 0.01, 1, 10
plt.xticks([0.001, 0.01, 0.1, 1, 10, 100], ['0.001', '0.01', '0.1', '1', '10', '100'])


plt.xlabel('Particle Size (mm)')
plt.ylabel('Percent Passing (%)')   
plt.grid(True, which="both", ls="--")
plt.tight_layout()
#save as svg
plt.savefig('Modeule2_GSD_plot.svg')
```
